# Linear Regression

Linear Regression: y = b + Xw. b = bias (offset). w = weight (slope). 
The bias controls the fixed ofset in our prediction. While the weight controls the proportional content in our prediction.

Linear Regression is a hypothesis that is controlled by two parameters weights and bias in order to find the best possible model to describer a data.

### Cost Function

Recall: we are solving a supervized learning task so we know the true value of the labels. Therefore, one can compare the value predicted by the hypthesis with the actual value of the label and calculate the error for each data point. 

These are called residuals and are calculated by subtracting the value of the prediction from the true value of the label. 

Note: in this definition a residual is positive if our hyphotesis underestimates the true price and negative if all the estimates? The goal is to find the total amount of being wrong. Therefore, we can define the total error as the sum of the absolute values of the residuals. 

The total error is one possible example of what is called the COST FUNCTION. A cost function returns a value once the hypotheses withheld the parameters B and W and the training data are set to fix values. 

Sometimes it is preferable to use another cost function called the mean square. This is calculated by taking the square of each residual summing all the squares and divided by the total number of data points in. 

Note that since this square is a positive function, the mean square will be large when the total error is large and small. When the total error is small in that sense they are somewhat equivalent. However, the mean square error will be much larger when individual differences are large because of the square operation. Additionally, the mean sqaure error is perfereable because it's smooth and is guaranteed to have a global minimun. 

Conclusion: Two types of cost function - The total error and the mean squared error. Also, for a given values of training data and hypotheses the cost only depends on the value of the parameters. 

```python
def mean_squared_error(y_true, y_pred):
    s = (y_true - y_pred)**2
    return s.mean()
```
