# Deep Learning Intro

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Shallow and Deep Networks

```python
# function: "make_moons"
# This is fake data. It does not represent anything. 
# The two data sets are not really separable with a line.
# Ultimate goal: The model need to be smart enough to accomodate for a boundary that goes around the shape.

# Scikit-learn = free ml library for python. It also supports NumPy and SciPy. 

from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.1, random_state=0)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.legend(['0', '1'])
```

```python
X.shape

# The input data includes thousand points
# Each point has two features
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and test. 
# Test size of 30%
# Random state so we get same result
# It sets the seed of the random number generator in order to get same random split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, Adam
```

### Shallow Model

```python
# shallow model = logistic regression
```

```python
# Build logistic regression model

model = Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid'))
model.compile(Adam(lr=0.05), 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, verbose=0)
```

```python
results = model.evaluate(X_test, y_test)
```

```python
results
```

```python
print("The Accuracy score on the Train set is:\t{:2f}".format(results[1]))
```

```python
def plot_decision_boundary(model, X, y):
    amin, bmin = X.min(axis=0) - 0.1
    amax, bmax = X.max(axis=0) + 0.1
    hticks = np.linspace(amin, amax, 101)
    vticks = np.linspace(bmin, bmax, 101)
    
    aa, bb = np.meshgrid(hticks, vticks)
    ab = np.c_[aa.ravel(), bb.ravel()]
    
    c = model.predict(ab)
    cc = c.reshape(aa.shape)

    plt.figure(figsize=(12, 8))
    plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
    plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
    plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
    plt.legend(['0', '1'])
    
plot_decision_boundary(model, X, y)
```

### Deep model

In deep learning, you want to build the models by essentially stacking layers of logistic regressions with many nodes or layers with different activation functions. 

1st - do a weighted avg of the inputs of the previous layer and then we apply nonlinear activation function. 

```python
# apply different activation functions
# activation function = converts an input signal of a node in a NN to an output sigal. 

model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='tanh')) #4 nodes
model.add(Dense(2, activation='tanh')) #2 nodes
model.add(Dense(1, activation='sigmoid')) # 1 node
model.compile(Adam(lr=0.05), 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, verbose=0)
```

```python
model.evaluate(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# check confusion matrix and the accuracy score

y_train_pred = model.predict_classes(X_train)
y_test_pred = model.predict_classes(X_test)

print("The Accuracy score on the Train set is:\t{:0.3f}".format(accuracy_score(y_train, y_train_pred)))
print("The Accuracy score on the Test set is:\t{:0.3f}".format(accuracy_score(y_test, y_test_pred)))
```

```python
plot_decision_boundary(model, X, y)
```

## Multiclass classification

### The Iris dataset

```python
df = pd.read_csv('../data/iris.csv')
```

```python
import seaborn as sns
sns.pairplot(df, hue="species")
```

```python
df.head()
```

```python
X = df.drop('species', axis=1)
X.head()
```

```python
target_names = df['species'].unique()
target_names
```

```python
target_dict = {n:i for i, n in enumerate(target_names)}
target_dict
```

```python
y= df['species'].map(target_dict)
y.head()
```

```python
from keras.utils.np_utils import to_categorical
```

```python
y_cat = to_categorical(y)
```

```python
y_cat[:10]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y_cat,
                                                    test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(3, input_shape=(4,), activation='softmax'))
model.compile(Adam(lr=0.1),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:5]
```

```python
y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_class, y_pred_class))
```

```python
confusion_matrix(y_test_class, y_pred_class)
```

The [Pima Indians dataset](https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes) is a very famous dataset distributed by UCI and originally collected from the National Institute of Diabetes and Digestive and Kidney Diseases. It contains data from clinical exams for women age 21 and above of Pima indian origins. The objective is to predict based on diagnostic measurements whether a patient has diabetes.

It has the following features:

- Pregnancies: Number of times pregnant
- Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- BloodPressure: Diastolic blood pressure (mm Hg)
- SkinThickness: Triceps skin fold thickness (mm)
- Insulin: 2-Hour serum insulin (mu U/ml)
- BMI: Body mass index (weight in kg/(height in m)^2)
- DiabetesPedigreeFunction: Diabetes pedigree function
- Age: Age (years)

The last colum is the outcome, and it is a binary variable.
